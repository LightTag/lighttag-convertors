```python
with open("/home/tal/Downloads/ner_prod_01_annotations.json") as f:
    ltjob = json.load(f)
```

```python
# We get back a dictionary mapping filenames to the actual content
annotated_gml_dict = lighttag_to_gml_pt.lighttag_job_to_pt_gml_dict(ltjob)
```

```python
csv_data =[]
#Viewing the annotations
for filename,gml_with_annotations in annotated_gml_dict.items():
    blob_path = f" gs://{bucket_name}/{filename}"
    with open(os.path.join("/tmp/jsonls",filename),'w') as f:
        json.dump(gml_with_annotations,f)
    csv_data.append({"jsonl_path":blob_path})
```

```python
import csv
keys = csv_data[0].keys()
with open("/tmp/nerlist.csv","w") as f:
    dict_writer = csv.DictWriter(f,keys)
    dict_writer.writerows(csv_data)
#     blob = bucket.blob("datalist.csv")
#     blob.upload_from_filename("/tmp/nerlist.csv")
```
